```python
infilename = "/home/ec2-user/store/c2gen/test/TaleOfTwoCities_pt09.c2cb"
indata = np.fromfile(infilename, dtype=np.uint8)
data = np.reshape(indata, (-1,16))  

for i in range(0,16):
  x = data[0:-2,i]  
  y = data[1:-1,i]

  e = data[0:-2,5]  
  c = []
  for dy in e:
    cy = float(dy)/64+0.5
    c.append([cy, cy/4, cy/4])
  plt.scatter(x,y, color=c, alpha=0.002, edgecolors='none')
  plt.grid(False)
  fn = "codec-plot-" + str(i) + ".png"
  plt.savefig(fn)  
  plt.show()
```
